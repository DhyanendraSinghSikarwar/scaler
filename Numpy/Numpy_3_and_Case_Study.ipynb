# Vectorization in NumPy

```python
ratings = np.array([4.2,3.5,4.2,4.9,4.2,4.6,4.3,3.2,2.4])
```

```python
ratings
```

```python
#rating>4.0 - High, low
#rating 4.5> high , 4- 4.5 medium, less than 4 low
```

```python
ratings >=4.0
```

```python
if ratings >=4.0:
  pass
```

```python
def categorize_rating(rating):
  if rating >=4.0:
    return "High"
  else:
    return "Low"
```

```python
categorize_rating(ratings)
```

```python
ratings
```

```python
np.vectorize(categorize_rating)
```

```python
vectorized_categorize_rating = np.vectorize(categorize_rating)
```

```python
vectorized_categorize_rating(ratings)
```

```python
def discount_cost(rate, cost):
  if rate>=4.0:
    return  cost *0.9
  else:
    return cost
```
